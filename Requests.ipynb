```python
import requests
emailurl='https://gist.githubusercontent.com/itsbalamurali/8231301/raw/c603eb4389ea6df78ea74c12f905c664ddabc55e/emails.txt'
resp=requests.get(emailurl)
content=resp.text
#content.split('\n')

eachlines=[]
emails=[]

for lines in content.split('\n'):
    if '@' in lines and '.' in lines:
        eachlines.append(lines.split(':')[1])
    
for email in eachlines:
    if ',' in email:
        result=email.split(',')
        emails.extend(result)
    else:
        emails.append(email)
        
emails
```
