```python
import requests

#url='https://ia800302.us.archive.org/20/items/romeojuliet01shak/romeojuliet01shak_djvu.txt'
url='https://ia800302.us.archive.org/10/items/draculabr00stokuoft/draculabr00stokuoft_djvu.txt'
resp=requests.get(url)
data=resp.text

# declaration
nospaces=[]
wordcount={}

# for loop to split into lines and collect non empty lines
for eachline in data.split('\n'):
     if eachline !='':
            nospaces.append(eachline.lower())

# for loop to split eachline into words and get the count
for line in nospaces:
#for word in nospaces.split(' '):
    for word in line.split(' '):
        if word not in wordcount:
            wordcount[word]=1
        else:
            wordcount[word]=wordcount[word]+1

# display the results
wordcount
#list(wordcount.values('sdition'))
```
